```python
import numpy as np
import pandas as pd
```

```python


# Load the dataset
df = pd.read_csv('a1_RestaurantReviews_HistoricDump.tsv', delimiter = '\t', quoting = 3)
```

```python
# data visualization
df.head()
```

```python
#data preprocessing
#data cleaning and removing stopwords
import re
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
#manipulating corpus
corpus=[]
for i in range(0, 900):
  review = re.sub('[^a-zA-Z]','',df['Review'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)

```

```python
#checking corpus 
corpus
```

```python
#data transformation
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(max_features = 1420)
```

```python
X = cv.fit_transform(corpus).toarray()
y = df.iloc[:, -1].values
```

```python
#Bow Dictionary for prediction
import pickle 
bow_path = 'c1_Bow_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

```python
#split of dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train ,y_test = train_test_split(X,y,test_size = 0.20,random_state=100)
```

```python
#Naive Bayes algorithm
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
#NB classifier for prediction purposes
import joblib
joblib.dump(classifier, 'c2_Classifier_Sentiment_Model')
```

```python
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test,y_pred)
```

```python
df['predicted_label'] = y_pred.tolist()
df.head()
```
